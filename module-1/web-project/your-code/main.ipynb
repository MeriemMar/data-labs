```python
n=10 # number of pages
names=[]
localization=[]
distance=[]
score_review=[]
experience=[]
score=[]
link=[]

for i in range(15,15*(67-1),15):
    
    page_response = requests.get(url%i)
    page_content = BeautifulSoup(page_response.content,'lxml') 
    
    names.extend([element.text.strip() for element in page_content.select('span.sr-hotel__name')])    
    localization.extend ([element.text.strip().split('\n')[0] for element in page_content.select('div.address a+a')])
    distance.extend([element.text.replace('from center','').strip() for element in page_content.select('div.address>span:first-of-type>span+span')])
    score_review.extend([element.text.strip() for element in page_content.select('div.bui-review-score__title')])
    experience.extend([element.text.strip().replace('\xa0', '') for element in page_content.select('div.bui-review-score div.bui-review-score__text')])
    score.extend([element.text.strip() for element in page_content.select('div.bui-review-score__badge')])
    link.extend(['https://www.booking.com/'+ str(element.get('href')).replace('\n','') for element in page_content.select('div.address>a+a')])
    
    
    
    
    
```

```python
col = ["Name", "Localization","Distance to center","Score","Experiences number", "Links"]
#data = list(zip(names,localization,distance,score,experience,link))
df = pd.DataFrame([names,localization,distance,score,experience,link]).T
df.columns=col
df
```

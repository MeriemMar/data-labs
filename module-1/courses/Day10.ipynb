# Outliers

- Detect outliers
- choose either to fix it 
         or drop it
        or leave it as it is

```python
df.boxplot(column='cylinders') #25 to 75 percentiles of data, median is the green , black barrs are acceptable values 
```

```python
df_desc= df.describe().T
```

```python
df_desc['IQR']= df_desc["75%"]-df_desc["25%"]
df_desc
```

```python
outliers= pd.DataFrame(columns= df.columns)
for col in df_desc.index:
    iqr= df_desc.at[col,'IQR']
    cutoff= iqr*1.5 # common value for the cutoffwe can use 1.5 or 1.75
    lower=df_desc.at[col,'25%']-cutoff
    upper=df_desc.at[col,'75%']+cutoff
    results= df[(df[col]<lower)|(df[col]>upper)].copy()
    results['outlier']=col 
    outliers=outliers.append(results)
outliers
```

# Lambda function

```python
# in a function we create and which disappear after we used it
data
```

```python
#if we want only verify this
data[data.JOB_LVL>100]
```

```python
#we create then a function
def is_high(x):
    return x>100

data[data["JOB_LVL"].apply(is_high)]

# create a lambda function
data[data["JOB_LVL"].apply(lambda x:x>100)]
```

```python
data[data.apply(lambda x:(x.JOB_LVL+x.JOB_ID)>100, axis=1)]
```

```python
square= lambda x:x**2
square(5)
# the same as
def square_fun(x):
    return x**2
square_fun(5)
```

```python
summ = lambda a,b: a+b
summ('a','b')
```

```python
div= lambda x,y: x/y if (y!=0) else 0

print(div(10,3))
print(div(10,0))
```

```python
def generate_range(lower):
    return lambda upper: range(lower,upper)
custom_range=generate_range(1)
print(custom_range(5))
```

```python
[x for x in custom_range(10)]
```

```python
[[x for x in custom_range(i)] for i in range(3,6)]
```

```python
[square(x) for x in range(6)]
```

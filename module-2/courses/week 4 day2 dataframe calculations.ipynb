# DataFrame Calculations

```python
import numpy as np
import pandas as pd
animals = pd.read_csv('https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/data/module-2/animals.csv')
```

```python
animals.head()
```

```python
animals.bodywt = animals.bodywt*1000
```

```python
animals.head(3)
```

```python
animals['brainratio'] = animals.brainwt/animals.bodywt
```

```python
animals.head()
```

```python
animals['reversecheck'] = animals.bodywt/animals.brainwt[animals.brainwt>50]
```

```python
animals
```

```python
animals['jazz'] = np.where(animals.brainwt>50, animals.bodywt/animals.brainwt,0)
```

```python
animals['average'] = animals.mean(axis=1)
animals
```

```python
vehicles.groupby(['Transmission'])
```

```python
vehicles.groupby(['Transmission'])['Highway MPG', 'City MPG', 'Combined MPG'].mean()
```

```python
vehicles.groupby(['Transmission'])['CO2 Emission Grams/Mile'].median()
```

```python
vehicles.groupby(['Fuel Type', 'Cylinders'])['CO2 Emission Grams/Mile'].median()
```

```python
from scipy import stats 
def agg_mode(x):
    return(stats.mode(x)[0])
```

```python
vehicles.groupby('Transmission')['Vehicle Class'].agg(agg_mode)
```

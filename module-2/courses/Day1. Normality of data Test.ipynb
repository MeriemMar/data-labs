```python
from scipy.stats import kstest
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame([np.random.normal(1,1,1000), np.random.normal(20,1,1000), np.random.normal(14,3,1000)]).T
df.columns=['a','b','c']
```

```python
df.head()
```

```python
#standardize the column

df.a=(df.a-df.a.mean())/df.a.std()
```

```python
kstest(df.a,'norm')

#we canot reject the hypothesis that the column is normal
```

```python
#standardize the whole dataframe
(df-df.mean())/df.std()
```

```python
for i in df:
    print(kstest(df[i],'norm'))
```
